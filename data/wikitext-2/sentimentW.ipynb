```python
df = pd.read_csv('./cvaw4.csv')
```

```python
sent = ' '.join(list(df['Word']))
```

```python
df['V_normalization'] = df['Valence_Mean'].apply(lambda x: round((x-1)/8,2))
```

```python
df['A_normalization'] = df['Arousal_Mean'].apply(lambda x: round((x-1)/8,2))
```
